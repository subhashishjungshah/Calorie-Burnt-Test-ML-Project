## CONVERTING TEXT DATA TO NUMERICAL DATA (CATEGORICAL DATA ENCODING)

```python
calories_Data.replace({"Gender":{'male':0,'female':1}},inplace=True) #inplace = true changes permanently
```

```python
calories_Data.head()
```

## SEPERATING FEATURES AND TARGET

```python
X = calories_Data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_Data['Calories']
```

```python
#print(X)
#print(Y)
```

## SPLITING DATA SET INTO TRAIN AND TEST


```python
#spliting the dataset into test and train. Test_size = 0.2 means that 20% of the total data is used for testing purpose.
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#checking different data shapes
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

```python
# XGBoost Regressor 
model = XGBRegressor()
```

```python
# Training the model with X_train
model.fit(X_train,Y_train)
```

```python
# Prediction on Test Data
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

## Mean Absolute Value

```python
#finding mean absolute error
mae = mean_absolute_error(Y_test,test_data_prediction)
print(mae)
```

```python
r2xgb = r2_score(Y_test, test_data_prediction)
print("R2 Score:", r2xgb)
```

```python
predictions = model.predict(X_test)
# Plot the predicted values against the true values
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("XGBoost Model")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model=LinearRegression()
lin_model.fit(X_train,Y_train)
pred_lin=lin_model.predict(X_test)
```

```python
mael =mean_absolute_error(Y_test,pred_lin)
```

```python
print("Mean Absolute Error= ", mael)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, pred_lin)
print("R2 Score:", r2)
```

```python
plt.scatter(Y_test, pred_lin)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression Model")
plt.show()
```
